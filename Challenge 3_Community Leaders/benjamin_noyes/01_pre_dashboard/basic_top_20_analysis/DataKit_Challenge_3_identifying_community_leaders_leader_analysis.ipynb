```python
leaders_gb = wefarm.groupby(["response_user_id"]
  ).agg({"response_id": "count"}
  ).sort_values(by=["response_id"], ascending=False)
leaders_user_id_lst = leaders_gb[:20].index
```

# niche

```python
niche = pd.read_parquet("/content/drive/MyDrive/DataKit_WeFarm/en_questions_cat_niche.parquet")
```

```python
niche.head(5)
```

```python
niche_user_id_df = pd.merge(niche, wefarm, on="question_id")
niche_user_id_df = niche_user_id_df.iloc[:, :72].drop_duplicates()
```

```python
niche_leaders_df = niche_user_id_df[niche_user_id_df["question_user_id"].isin(leaders_user_id_lst)]
niche_nonleaders_df = niche_user_id_df[~niche_user_id_df["question_user_id"].isin(leaders_user_id_lst)]
```

```python
len(niche_leaders_df)+len(niche_nonleaders_df) == len(niche)
```

```python
niche_leaders_df[:5]
```

```python
def sort_niche_df(df, collst):

  keyword_sums = df[collst].sum().sort_values()
  sums_df = keyword_sums.reset_index()
  sums_df.columns = ["col", "sum"]

  sums_df["prefix"] = [c.split("_")[0] for c in sums_df["col"]]

  sums_df = sums_df.sort_values(["prefix", "sum"], ascending=[True, False])
  sums_df["sum_pct"] = sums_df["sum"] / sums_df["sum"].sum()

  return sums_df
```

```python
category_collst = niche_leaders_df.columns[3:71]

niche_leaders_sum_df = sort_niche_df(niche_leaders_df, category_collst)
niche_nonleaders_sum_df = sort_niche_df(niche_nonleaders_df, category_collst)
```

```python
niche_leaders_sum_df
```

```python
import plotly.graph_objects as go

fig = go.Figure()

color_lst = ["#74C476", "#DD98DA"]
name_lst = ["leaders", "nonleaders"]
dfs_lst = [niche_leaders_sum_df, niche_nonleaders_sum_df]

for idx, df in enumerate(dfs_lst):
  fig.add_trace(
    go.Bar(
      y=df["sum_pct"],
      x=df["col"],
      marker={
        "color": df["sum_pct"],
        "color": color_lst[idx]
      },
      name=name_lst[idx],
    )
  )

fig.update_layout({
  "title": '"leaders" ask more questions about livestock, fewer about crops, disease, market, and seasonality'
})
```

# broad

```python
broad = pd.read_parquet("/content/drive/MyDrive/DataKit_WeFarm/en_questions_cat_broad.parquet")
```

```python
broad.head(5)
```

```python
broad_user_id_df = pd.merge(broad, wefarm, on="question_id").iloc[:, :9].drop_duplicates()
```

```python
broad_leaders_df = broad_user_id_df[broad_user_id_df["question_user_id"].isin(leaders_user_id_lst)]
broad_nonleaders_df = broad_user_id_df[~broad_user_id_df["question_user_id"].isin(leaders_user_id_lst)]
```

```python
len(broad_leaders_df)+len(broad_nonleaders_df) == len(broad)
```

```python
broad_leaders_df[:5]
```

```python
def sum_counts(df):
  keyword_collst = df.columns[3:8]
  sum_arr = df[keyword_collst].sum().values
  sum_perc_arr = sum_arr / sum_arr.sum()
  return pd.DataFrame({
    "col": keyword_collst,
    "sum": sum_arr,
    "sum_pct": sum_perc_arr
  })
```

```python
broad_leaders_sum_df = sum_counts(broad_leaders_df)
broad_nonleaders_sum_df = sum_counts(broad_nonleaders_df)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

color_lst = ["#74C476", "#DD98DA"]
name_lst = ["leaders", "nonleaders"]
dfs_lst = [broad_leaders_sum_df, broad_nonleaders_sum_df]

for idx, df in enumerate(dfs_lst):
  fig.add_trace(
    go.Bar(
      x=df["col"],
      y=df["sum_pct"],
      marker={
        "color": color_lst[idx]
      },
      name=name_lst[idx],
    )
  )

fig.update_layout({
  "title": '"leaders" ask more questions about livestock'
})
```
