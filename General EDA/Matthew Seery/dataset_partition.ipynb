# DATASET PARTITIONING

```python
import pandas as pd
```

```python
chunks = pd.read_csv('../data/cleaned_updated_data.csv',
                     dtype={'question_user_gender': str, 'response_user_gender': str}, # Removes mixed dtypes error message
                     chunksize=100000)
df = pd.DataFrame()
```

```python
for chunk in chunks:
    df = pd.concat([df,chunk], axis=0)
```

```python
df.info()
```

```python
columns = list(df.columns)
for column in columns:
    missing = 'len(df[df.'+column+'.isnull() == True])'
    print(f'{column} has {eval(missing)} missing values')
```

```python
df_question_topic_valid = df[~df.question_topic.isnull()]
df_question_topic_null = df[df.question_topic.isnull()]

df_question_topic_valid.to_csv('../data/question_topic_valid.csv', index=False)
df_question_topic_null.to_csv('../data/question_topic_null.csv', index=False)
```

```python
df_question_user_country_code_ke = df[df.question_user_country_code == 'ke']
df_question_user_country_code_ug = df[df.question_user_country_code == 'ug']
df_question_user_country_code_tz = df[df.question_user_country_code == 'tz']
df_question_user_country_code_gb = df[df.question_user_country_code == 'gb']

df_question_user_country_code_ke.to_csv('../data/question_user_country_code_ke.csv', index=False)
df_question_user_country_code_ug.to_csv('../data/question_user_country_code_ug.csv', index=False)
df_question_user_country_code_tz.to_csv('../data/question_user_country_code_tz.csv', index=False)
df_question_user_country_code_gb.to_csv('../data/question_user_country_code_gb.csv', index=False)
```

```python
df_question_language_eng = df[df.question_language == 'eng']
df_question_language_swa = df[df.question_language == 'swa']
df_question_language_nyn = df[df.question_language == 'nyn']
df_question_language_lug = df[df.question_language == 'lug']

df_question_language_eng.to_csv('../data/question_language_eng.csv', index=False)
df_question_language_swa.to_csv('../data/question_language_swa.csv', index=False)
df_question_language_nyn.to_csv('../data/question_language_nyn.csv', index=False)
df_question_language_lug.to_csv('../data/question_language_lug.csv', index=False)
```

```python
df_question_user_status_live = df[df.question_user_status == 'live']
df_question_user_status_zombie = df[df.question_user_status == 'zombie']
df_question_user_status_destroyed = df[df.question_user_status == 'destroyed']
df_question_user_status_blocked = df[df.question_user_status == 'blocked']

df_question_user_status_live.to_csv('../data/question_user_status_live.csv', index=False)
df_question_user_status_zombie.to_csv('../data/question_user_status_zombie.csv', index=False)
df_question_user_status_destroyed.to_csv('../data/question_user_status_destroyed.csv', index=False)
df_question_user_status_blocked.to_csv('../data/question_user_status_blocked.csv', index=False)
```
